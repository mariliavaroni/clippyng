```python
import requests
import datetime
from bs4 import BeautifulSoup
nivelMaximo = 2
qtdTotal = 0
qtdEncontrado = 0
busca = ""
def request(url):
    try:
        return requests.get(url).text
    except Exception as e:
        return ""
        
def getURLLinks(url):
    param = {}
    result = request(url)
    soup = BeautifulSoup(result, "html5lib")
    url = set()
    for links in soup.find_all('a'):
         url.add(links.get('href'))
    return url
def getURLText(url,palavra):
    html = request(url)
    soup = BeautifulSoup(html, 'html.parser')
    # tira todos os elementos "script" e "style" 
    for script in soup(["script", "style"]):
        script.extract()
    
    texto = soup.get_text()
    texto = texto.lower()
    
    if palavra in texto:
        trechos(texto,palavra)
        return True
    else:
        return False
    
def trechos(texto,busca):
    listaTrechos = []
    indice = 0
    palavras = texto.split(" ")
    for palavra in palavras:
        if busca.strip() in palavra:
            frase = ""
            for x in range(-5, 5):
                try:
                    frase += " " + (palavras[indice+x])
                except Exception as e:
                    continue
            frase = frase.replace("\n","").replace("\t","").replace("\r","")
            listaTrechos.append(frase) 
        indice = indice+1
    i = 1;
    for trecho in listaTrechos:
        print(f"{i}){trecho}")
        i += 1
    
def limpa(url):
    url = url.replace(';', '')
    return url
    
def pesquisa(sites, palavra, nivel):
    global qtdTotal
    global qtdEncontrado
    
    if nivel <= nivelMaximo:
        for site in sites:
            if site:
                site = limpa(site)
                if getURLText(site,palavra):
                    qtdEncontrado = qtdEncontrado+1
                    print(f"\o/ Encontrado em {site} \n")
                    
                qtdTotal = qtdTotal+1
                
                #sub sites:
                pesquisa(getURLLinks(site), palavra, nivel+1)
def exibeResultado():
    print(f"A palavra '{busca}' foi procurada em {qtdTotal} sites e foi encontrada em {qtdEncontrado}.")
    
def clippyng(sites, palavra):
    global busca
    print("Inicio: "+str(datetime.datetime.now()))
    busca = palavra
    sites = sites.split(";")
    pesquisa(sites, palavra, 1)
    exibeResultado()
    print("Final: "+str(datetime.datetime.now()))
    
def manual():
    sites = input ("Entre com uma lista de sites separados por ponto e virgula (;): ")
    palavra = input ("Entre com a palavra que deseja buscar: ")
    clippyng(sites,palavra)
    
def teste():
    sites = "http://g1.globo.com/; http://www.folha.uol.com.br/; https://www.r7.com/;"
    palavra = " trÃ¡fico "
    clippyng(sites,palavra)
    
    
manual()
#teste()
```
